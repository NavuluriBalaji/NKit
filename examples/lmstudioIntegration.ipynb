```python
from NKit.llms import LMStudioLLM
from NKit.observer import LiveObserver
from NKit import Agent

# 1. Initialize the observer so you can watch what the agent is thinking in real-time
observer = LiveObserver()

# 2. Connects to whatever model you currently have loaded in LM Studio GUI!
local_llm = LMStudioLLM() 

# 3. Create the agent
agent = Agent(
    llm=local_llm, 
    observer=observer
)

# 4. Bind a print function so you can see the thoughts as they stream out
@observer.on("agent.reasoning")
def print_thoughts(event):
    print(f"\n[THOUGHT]: {event.get('thought', '')}")

print("LMStudio Agent Successfully Initialized!")
```

```python
from NKit.llms import LMStudioLLM

# Pass the exact tuning URL + Port your LM Studio Local Inference Server is running on!
local_llm = LMStudioLLM(
    model="nvidia/nemotron-3-nano-4b", 
    base_url="http://localhost:312", # <-- Change this to any port or IP!
    temperature=0.7
)

agent = Agent(llm=local_llm, observer=observer)
```
